# With EXtra-geom
EXtra-geom is a Python library to describe the physical layout of multi-module detectors at European XFEL, and to assemble complete detector images. See https://extra-geom.readthedocs.io/en/latest/

![extra_geom geometry](https://extra-geom.readthedocs.io/en/latest/_images/agipd_layout.png)

```python
from extra_geom import AGIPD_1MGeometry
import sgeom as sg
import numpy as np
import matplotlib.pyplot as plt
```

```python
geom = AGIPD_1MGeometry.from_crystfel_geom("AGIPD.geom")
geom.inspect(axis_units='px')
```

## Standard transmission geometry

```python
clen = 0.1222
pixel_size = geom.pixel_size
pixel_positions = geom.get_pixel_positions()
sgeom = sg.sgeom.from_extra_geom_pixel_positions(pixel_positions, clen, pixel_size)
sgeom.info()
```

```python
print(sgeom.qspace.shape)
geom.plot_data(sgeom.qspace*1e-10*2*np.pi)
plt.title(r"|$\vec{q}$| ($\rm \AA^{-1}$)")
```

```python
geom.plot_data(sgeom.reciprocal[..., 0]*1e-10*2*np.pi)
plt.title(r"qx ($\rm \AA^{-1}$)")
```

```python
geom.plot_data(sgeom.reciprocal[..., 2]*1e-10*2*np.pi)
plt.title(r"qy ($\rm \AA^{-1}$)")
```

```python
geom.plot_data(sgeom.reciprocal[..., 1]*1e-10*2*np.pi)
plt.title(r"qz ($\rm \AA^{-1}$)")
```

## Tilted incident X-ray

```python
clen = 0.1222
pixel_size = geom.pixel_size
beam_vec_lab = (0.3,0.3,1)
clen = 0.1222

pixel_positions = geom.get_pixel_positions()
sgeom = sg.sgeom.from_extra_geom_pixel_positions(pixel_positions, clen, pixel_size, beam_vec_lab=beam_vec_lab)
sgeom.info()
```

```python
print(sgeom.qspace.shape)
geom.plot_data(sgeom.qspace*1e-10*2*np.pi)
plt.title(r"|$\vec{q}$| ($\rm \AA^{-1}$)")
```
