# Geometry setting

npixels (tuple[int]): Number of pixels in (vertical, horizontal) directions, e.g. (128, 128).

poni (tuple[float]): Point Of Normal Incidence coordinates in the detector system in pixel unit ([vertical, horizontal]) from the lower left.

clen (float): Distance between the sample and the PONI in meters.

pixel_size (tuple[float]): The size of a single pixel ([height, width]) in meters.

X1 (tuple, optional): The X1 axis in lab/beam system toward the ceiling, expressed in sample coordinates. Defaults to sample (0, 0, 1).  
X2 (tuple, optional): The X2 axis in lab/beam system transverse to the incident beam, expressed in sample coordinates. Defaults to sample (1, 0, 0).  
X3 (tuple, optional): The X3 axis in lab/beam system along the incident beam toward the detector in sample coordinates. It defines the beam direction with respect to the sample system. Defaults to sample (0, 1, 0).  

rot (tuple, optional): [rot1, rot2, rot3] in radians - see pyFAI geometry (https://pyfai.readthedocs.io/en/master/geometry_conversion.html#geometry-definition-of-pyfai for details. Defaults to (0, 0, 0).

wavelength (float): The wavelength of the beam in meters. Defaults to 1.33e-10 (12.398/9.3 KeV).

# Direct beam at center

```python
geom = sgeom.from_pixel_number([80, 80], 0.13, [40,40], 1200e-6, wavelength = 2.5e-10)
geom.info()
```

### Here shows the position map of these pixels in detector coordinate.

```python
plt.figure()
plt.imshow(geom.pixels_lab[...,0], origin="lower")
plt.colorbar()
plt.title("Vertical direction")
plt.figure()
plt.imshow(geom.pixels_lab[...,1], origin="lower")
plt.colorbar()
plt.title("Horizontal direction")
```

## Reciprocal space 

```python
plt.figure()
plt.imshow(geom.reciprocal[...,0]*1e-10, origin="lower")
plt.colorbar()
plt.title(r"qx ($\rm \AA^{-1}$)")
plt.figure()
plt.imshow(geom.reciprocal[...,2]*1e-10, origin="lower")
plt.colorbar()
plt.title(r"qz ($\rm \AA^{-1}$)")
plt.figure()
plt.imshow(geom.reciprocal[...,1]*1e-10, origin="lower")
plt.colorbar()
plt.title(r"qy ($\rm \AA^{-1}$)")
```

```python
plt.figure()
plt.imshow(geom.qspace*1e-10, origin="lower")
plt.colorbar()
plt.title(r"|$\vec{q}$| ($\rm \AA^{-1}$)")
```

# Rotate detector

## Rotate detector up

```python
geom = sgeom.from_pixel_number([80, 80], 0.13, [40,40], 1200e-6, wavelength = 2.5e-10, rot=[0, -np.deg2rad(16.41), 0])
geom.info()
```

```python
plt.figure()
plt.imshow(geom.reciprocal[...,0]*1e-10, origin="lower")
plt.colorbar()
plt.title(r"qx ($\rm \AA^{-1}$)")
plt.figure()
plt.imshow(geom.reciprocal[...,2]*1e-10, origin="lower")
plt.colorbar()
plt.title(r"qz ($\rm \AA^{-1}$)")
plt.figure()
plt.imshow(geom.reciprocal[...,1]*1e-10, origin="lower")
plt.colorbar()
plt.title(r"qy ($\rm \AA^{-1}$)")
plt.figure()
plt.imshow(geom.qspace*1e-10, origin="lower")
plt.colorbar()
plt.title(r"|$\vec{q}$| ($\rm \AA^{-1}$)")
```

## And rotate detector left

```python
geom = sgeom.from_pixel_number([80, 80], 0.13, [40,40], 1200e-6, wavelength = 2.5e-10, rot=[-np.deg2rad(16.41), -np.deg2rad(16.41), 0])
geom.info()
```
