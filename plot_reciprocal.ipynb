```python
import h5py
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
# fn = "/home/juncheng/GPFS/exfel/data/user/juncheng/hydratedProject/src/controller/3Dvolume/reciprocal_volume.h5"
fn = "reciprocal_volume.h5"
with h5py.File(fn, "r") as h5:
    volume = h5["volume"][()]
    q_map = h5["q_map"][()]*1e10
```

```python
my_slice = volume[:,:,109//2]
```

```python
plt.figure(figsize=(8,8))
plt.imshow(my_slice, norm=colors.LogNorm(), extent=(q_map[0,:,0,1].min(), q_map[0,:,0,1].max(), q_map[:,0,0,0].min(), q_map[:,0,0,0].max()))
plt.xlabel(r"$q_y$",fontsize=20)
plt.ylabel(r"$q_x$",fontsize=20)
```
