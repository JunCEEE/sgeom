```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
# fn = "/home/juncheng/GPFS/exfel/data/user/juncheng/hydratedProject/src/controller/3Dvolume/reciprocal_volume.h5"
fn = "reciprocal_volume.h5"
with h5py.File(fn, "r") as h5:
    volume = h5["volume"][()]
    q_map = h5["q_map"][()]*1e10
```

```python
qx = q_map[:,0,0,0]
qy = q_map[0,:,0,1]
qz = q_map[0,0,:,2]
```

```python
my_slice = volume[:,109//2,:]
my_slice = np.moveaxis(my_slice, 0, -1)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(my_slice, norm=colors.LogNorm(), extent=(qx.min(), qx.max(), qz.min(), qz.max()))
plt.xlabel(r"$q_y$",fontsize=20)
plt.ylabel(r"$q_x$",fontsize=20)
```
